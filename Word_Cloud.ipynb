```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
nltk.download('stopwords')
from nltk.corpus import stopwords
l = stopwords.words('english')
from datascience import *
```

```python
"""
Saving the word cloud as a vector version instead
"""
posttable = Table().read_table('posttable')
postmerged = ''
for x in posttable.column('posts'):
    postmerged = postmerged+' '+x
    
text2 = postmerged
wordcloud = WordCloud(width=2500, height=2500, margin=0,stopwords=l, collocations=False, background_color = 'white').generate(text2)
plt.figure(figsize = (50, 50))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('wordcloud2.eps', format = 'eps', dpi=50)
plt.show()
```
