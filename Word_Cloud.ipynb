```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
nltk.download('stopwords')
from nltk.corpus import stopwords
l = stopwords.words('english')
from datascience import *
```

```python
posttable = Table().read_table('posttable')
postmerged = ''
for x in posttable.column('posts'):
    postmerged = postmerged+' '+x
    
text2 = postmerged
wordcloud = WordCloud(width=800, height=800, margin=0,stopwords=l, collocations=False).generate(text2)
plt.figure(figsize = (10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
